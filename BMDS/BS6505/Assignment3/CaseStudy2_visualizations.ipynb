```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取数据
data = pd.read_csv("week6_data_CS2.txt", sep="\t")
data.head()
```

```python

# 箱线图 + 小提琴图
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 箱线图
sns.boxplot(x="Status", y="Score", data=data, ax=axes[0], palette="Set2")
axes[0].set_title("IQ Gain by Group (Boxplot)")
axes[0].set_xlabel("Group")
axes[0].set_ylabel("IQ Gain")

# 小提琴图
sns.violinplot(x="Status", y="Score", data=data, ax=axes[1], palette="Set2", inner="box")
axes[1].set_title("IQ Gain by Group (Violin Plot)")
axes[1].set_xlabel("Group")
axes[1].set_ylabel("IQ Gain")

plt.suptitle("Case Study 2: IQ Gain Comparison", fontsize=14)
plt.tight_layout()
plt.show()
```

```python

# 两组叠加的分布图 (Histogram + KDE)
plt.figure(figsize=(10,6))

sns.histplot(data=data, x="Score", hue="Status", element="step",
             stat="density", common_norm=False, bins=12, kde=True)

plt.title("Distribution of IQ Gains by Group")
plt.xlabel("IQ Gain")
plt.ylabel("Density")
plt.legend(title="Group")
plt.show()
```

```python

# 分别绘制两组单独的分布图
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Identified group
sns.histplot(data[data["Status"]=="Identified children"]["Score"], 
             bins=10, kde=True, ax=axes[0], color="skyblue")
axes[0].set_title("Identified Students: Distribution of IQ Gain")
axes[0].set_xlabel("IQ Gain")
axes[0].set_ylabel("Density")

# Not identified group
sns.histplot(data[data["Status"]=="Not identified children"]["Score"], 
             bins=10, kde=True, ax=axes[1], color="salmon")
axes[1].set_title("Not Identified Students: Distribution of IQ Gain")
axes[1].set_xlabel("IQ Gain")
axes[1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```
